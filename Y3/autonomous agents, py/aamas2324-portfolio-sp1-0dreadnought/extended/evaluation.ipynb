```python
import os
import sys
sys.path.append(os.path.abspath('..'))
from model import Farm
import matplotlib.pyplot as pl

field = Farm(npickers=2,nscouts=2, mode=True, width=9)
fieldf = Farm(npickers=2,nscouts=2, mode=False, width=9)

for i in range (1000):
    field.step()
    fieldf.step()

field_df = field.datacollector.get_model_vars_dataframe()
fieldf_df = fieldf.datacollector.get_model_vars_dataframe()
```

To evaluate the simulation I chose Total berries collected and Avg time it takes to
collect 1 berry as KPIs.
Bot normal mode and extended were launched for 1000 steps each. The results are shown below:
Extended operation mode not just collected much more than normal mode, but also does so more frequiently

```python
pl.figure()
pl.subplot(211)
pl.title("Simulation Statiscs")
pl.plot(field_df.score)
pl.plot(field_df.time)
pl.subplot(212)
pl.plot(fieldf_df.score)
pl.plot(fieldf_df.time)
pl.xlabel("Time (Steps)")
```
